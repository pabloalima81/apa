```
def Fib_I (position):
  n=position
  if (position == 1):
    return 0
  else:
    if (position == 2):
      return 1;
    else :
      first = 0
      second = 1
      print("1 ;",first,";",time.time() )
      print("2 ;",second,";",time.time())
      for count in range(2,n):
        inicial = time.time()
        value = first + second
        first = second
        second = value
        count += 1
        print(count,";" ,value, ";", time.time() - inicial)
```
